## 作業目標

* 根據範例，取出三天前下午三點到五點的新聞

## 根據範例，取出今天所有的新聞

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from datetime import datetime

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")

import time
for i in range(40):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    
soup = BeautifulSoup(browser.page_source, 'html.parser')

news = {}
for d in soup.find(class_="part_list_2").find_all('h3'):
    newsDate = d.find(class_="date").text
    newsLine = d.find_all('a')[-1].text
    news[newsDate] = "{0} {1}".format(newsDate, newsLine)
    #print(d.find(class_="date").text, d.find_all('a')[-1].text)
    
# 關閉瀏覽器
browser.quit();
```

## 取出今天下午三點到五點的新聞

```python
from datetime import datetime
today3 = datetime.strptime("2019/12/23 15:00", "%Y/%m/%d %H:%M")
today5 = datetime.strptime("2019/12/23 17:00", "%Y/%m/%d %H:%M")

for key in news.keys():
    d = datetime.strptime(key, "%Y/%m/%d %H:%M")
    if d < today5 and d > today3:
        print(news[key])
```

## 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium.webdriver.support.ui import Select

browser = webdriver.Chrome(executable_path='chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")

selectSite = Select(browser.find_element_by_id("selD"))
selectSite.select_by_value('20')

browser.find_element_by_id('button').click()

import time
for i in range(50):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")
    
soup = BeautifulSoup(browser.page_source, 'html.parser')

news2 = {}
for d in soup.find(class_="part_list_2").find_all('h3'):
    newsDate = d.find(class_="date").text
    newsLine = d.find_all('a')[-1].text
    news2[newsDate] = "{0} {1}".format(newsDate, newsLine)
    #print(d.find(class_="date").text, d.find_all('a')[-1].text)
    
# 關閉瀏覽器
browser.quit();

ThreeDay3 = datetime.strptime("2019/12/20 15:00", "%Y/%m/%d %H:%M")
ThreeDay5 = datetime.strptime("2019/12/20 17:00", "%Y/%m/%d %H:%M")

for key in news2.keys():
    d = datetime.strptime(key, "%Y/%m/%d %H:%M")
    if d < ThreeDay5 and d > ThreeDay3:
        print(news2[key])
```
