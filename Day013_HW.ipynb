## 作業目標

根據範例 ，完成以下問題：

* ① 印出最新文章的「作者」「標題」「時間」
* ② 印出第一頁所有文章的「作者」「標題」「時間」


### ① 印出最新文章的「作者」「標題」「時間」

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.ptt.cc/bbs/NBA/index.html'
r = requests.get(url)
soup = BeautifulSoup(r.text, "html.parser")
soup
```

```python
d = soup.find_all(class_="title")[0]
latest = BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href']).text, "html.parser")

for t in latest.find_all("div", {"class" : "article-metaline"}):
    print('{0}:{1}'.format(t.find(class_='article-meta-tag').text, t.find(class_='article-meta-value').text))
```

### ② 印出第一頁所有文章的「作者」「標題」「時間」

```python
for d in soup.find_all(class_="title"):
    latest = BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href']).text, "html.parser")
    for t in latest.find_all("div", {"class" : "article-metaline"}):
        print('{0}:{1}'.format(t.find(class_='article-meta-tag').text, t.find(class_='article-meta-value').text))
```

### ③ 試著爬爬看其他版的文章

```python
url = 'https://www.ptt.cc/bbs/ScenicPhoto/index.html'
r = requests.get(url)
soup = BeautifulSoup(r.text, "html.parser")

for d in soup.find_all(class_="title"):
    latest = BeautifulSoup(requests.get('https://www.ptt.cc'+d.find('a')['href']).text, "html.parser")
    for t in latest.find_all("div", {"class" : "article-metaline"}):
        print('{0}:{1}'.format(t.find(class_='article-meta-tag').text, t.find(class_='article-meta-value').text))
```
